```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
# Load CSV file into the df
krypto_df = pd.read_csv('crypto_data.csv')
krypto_df
```

```python
# IsTrading is true
krypto_df = krypto_df.loc[krypto_df['IsTrading'] == True]
```

```python
krypto_df.dropna()
```

```python
krypto_df = krypto_df.loc[krypto_df['TotalCoinsMined'] > 0]
krypto_df
```

```python
krypto_df = krypto_df.drop(columns=['CoinName','IsTrading','Unnamed: 0'])
```

```python
X = pd.get_dummies(krypto_df, columns=['Algorithm', 'ProofType'])
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=0.99)
krypto_pca = pca.fit_transform(X_scaled)
```

```python
krypto_pca
```

```python
krypto_pca = pd.DataFrame(krypto_pca)
```

```python
krypto_pca
```

```python
tsne = TSNE(learning_rate=35)
```

```python
tsne_features = tsne.fit_transform(krypto_pca)
tsne_features.shape
```

```python
x = tsne_features[:,0]
y = tsne_features[:,1]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
inertia = []
k = list(range(1, 11))
```

```python
for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(krypto_pca)
    inertia.append(km.inertia_)
```

```python
plot_data = {"k": k, "inertia": inertia}
plot_df = pd.DataFrame(plot_data)
plt.plot(plot_df['k'], plot_df['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
